- An image is worth of a thousand of words

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# A = np.array(range(0,10))
# A = np.arange(10)
A = np.linspace(1,10,100)
```

```python
A
```

## Create a one-dimensional array and plot it

```python
plt.plot(A)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("A simple line chart")
plt.grid()
```

## Create two curves and save them into an image file

```python
fig = plt.figure()
plt.plot(np.sin(A), '-')
plt.plot(np.cos(A), '--')
fig.savefig('myTrigChart.png')
```

```python
plt.figure()
plt.subplot(2,1,1) # rows, columns, panel number
plt.ylim(-2,2)
plt.plot(np.sin(A),label='sin(x)')
plt.legend()
plt.subplot(2,1,2)
plt.plot(np.cos(A),label='cos(x)')
plt.legend()
```

```python
fig = plt.figure()
ax = plt.axes()
plt.xlim(0,100)
plt.ylim(-2,2)
plt.plot(np.sin(A))
```
