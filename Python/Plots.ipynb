# Matplotlib

- An image is worth of a thousand of words

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# A = np.array(range(0,10))
# A = np.arange(10)
A = np.linspace(0,10,100)
```

```python
A
```

## Create a one-dimensional array and plot it

```python
plt.plot(A,'--o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("A simple line chart")
plt.grid()
```

```python
x = np.arange(10)
plt.scatter(x,A)
```

## Create two curves and save them into an image file

```python
fig = plt.figure()
plt.plot(np.sin(A), '-')
plt.plot(np.cos(A), '--')
fig.savefig('myTrigChart.png')
```

```python
plt.figure()
plt.subplot(2,1,1) # rows, columns, panel number
plt.ylim(-2,2)
plt.plot(np.sin(A),'--o',label='sin(x)')
plt.legend()
plt.subplot(2,1,2)
plt.ylim(-2,2)
plt.plot(np.cos(A),'-',label='cos(x)')
plt.legend()
```

```python
fig = plt.figure()
ax = plt.axes()
plt.xlim(0,100)
plt.ylim(-2,2)
plt.plot(np.sin(A))
```

# 3D plots

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')
z = np.linspace(0,150,1000)
x = np.sin(z)
y = np.cos(z)
ax.plot3D(x,y,z)
```

```python
x = np.arange(1, 100)
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(np.log(x))
plt.subplot(2,2,2)
plt.plot(np.exp(x))
plt.subplot(2,2,3)
plt.plot(x**2)
plt.subplot(2,2,4)
plt.plot(np.sqrt(x))
```
