```python
A = np.zeros([9,9])
A[0,:]=255
A[-1,:]=255
A[:,0]=255
A[:,-1]=255
```

```python
plt.imshow(A, cmap='gray')
```

```python
# How to blur the image? 
A[1,0] = (A[0,0]+A[2,0])/2
B = np.zeros([9,9])
B = np.copy(A)
```

```python
def myBlur(img):
    B = np.empty_like(img)
    for k in range(100):
        for i in range(1, len(A)-1):
            for j in range(1, len(A[0])-1):
                B[i,j] = (img[i-1, j] +img[i+1, j]+ img[i, j-1] +img[i, j+1])/4
        img = np.copy(B)
    return img
```

```python
#plt.imshow(A, cmap='gray')
B = myBlur(A)
plt.imshow(B, cmap='gray')
```
